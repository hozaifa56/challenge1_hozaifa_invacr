```python
service = Service(executable_path="chromedriver.exe")
driver = webdriver.Chrome(service=service)

base_url = 'https://www.bewakoof.com/men-clothing?category=t-shirt'
driver.get(base_url)

time.sleep(10)


def scroll_to_bottom(driver):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(3)  


last_height = driver.execute_script("return document.body.scrollHeight")
while True:
    scroll_to_bottom(driver)
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height
    
    
tshirt_detail=[]

tshirt_series=driver.find_elements(By.CSS_SELECTOR, 'h3.brand-name.undefined')
tshirt_title=driver.find_elements(By.CSS_SELECTOR, 'h2.clr-shade4.h3-p-name.undefined.false')
tshirt_mrp=driver.find_elements(By.CSS_SELECTOR, 'div.actualPriceText.clr-shade5')
tshirt_sp=driver.find_elements(By.CSS_SELECTOR, 'div.discountedPriceText.clr-p-black.false')
tshirt_rating=driver.find_elements(By.CSS_SELECTOR, 'span.clr-shade-3')
```

```python
for i in range(len(tshirt_title)):
    series=tshirt_series[i].text
    title = tshirt_title[i].text
    mrp = tshirt_mrp[i].text
    sp=tshirt_sp[i].text
    if i < len(tshirt_rating):
        rating = tshirt_rating[i].text
    else:
        rating = None

    tshirt_info={
        'Series':series,
        'Title':title,
        'MRP':mrp,
        'SP':sp,
        'Rating':rating
    }
    tshirt_detail.append(tshirt_info)

df=pd.DataFrame(tshirt_detail)
df.to_csv('bewakood_tshirt_info.csv')
```

# Rare Rabbit

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time

# Path to your ChromeDriver executable
service = Service(executable_path="chromedriver.exe")

# Create a new instance of the Chrome driver
driver = webdriver.Chrome(service=service)

# Open the page
base_url = 'https://www.flipkart.com/mens-tshirts/rare-rabbit~brand/pr?sid=clo%2Cash%2Cank%2Cedy&p%5B%5D=facets.brand%255B%255D%3DRARE%2BRABBIT'
tshirt_details = []

for page in range(1,11): 
    url = f"{base_url}&page={page}"  
    driver.get(url)

    titles = driver.find_elements(By.CSS_SELECTOR, 'a.WKTcLC.BwBZTg')
    price_elements = driver.find_elements(By.CSS_SELECTOR, 'div.Nx9bqj')

    mrp_elements = driver.find_elements(By.CSS_SELECTOR, 'div.yRaY8j')



    for i in range(len(titles)):
        title = titles[i].text
        price = price_elements[i].text.split('\n')[0]
        mrp = mrp_elements[i].text
        
        tshirt_info = {
            "Title": title,
            "SP": price,
            "MRP":mrp,
        }

        tshirt_details.append(tshirt_info)

df=pd.DataFrame(tshirt_details)
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import pandas as pd

# Path to your ChromeDriver executable
service = Service(executable_path="chromedriver.exe")

# Create a new instance of the Chrome driver
driver = webdriver.Chrome(service=service)

# Open Bewakoof T-shirt page
base_url = 'https://www.marchtee.com/classics'
driver.get(base_url)

# Wait for the page to load initially
time.sleep(30)

tshirt_detail = []

# Extract product details
tshirt_elements = driver.find_elements(By.CSS_SELECTOR, 'div.product-info')

for element in tshirt_elements:
    try:
        title = element.find_element(By.CSS_SELECTOR, 'p.title.footnote').text
    except:
        title = None
    try:
        color = element.find_elements(By.CSS_SELECTOR, 'p.footnote')[1].text
    except:
        color = None

    try:
        mrp = element.find_element(By.CSS_SELECTOR, 'span.price.original.strikethrough').text
    except:
        mrp = None

    try:
        sp = element.find_element(By.CSS_SELECTOR, 'span.price.discounted').text
    except:
        sp = element.find_element(By.CSS_SELECTOR, 'span.price.discounted').text

    tshirt_info = {
        'Title': title,
        'Color': color,
        'MRP': mrp,
        'SP': sp
    }
    tshirt_detail.append(tshirt_info)

# Convert the list of dictionaries to a DataFrame
df = pd.DataFrame(tshirt_detail)

# Save the DataFrame to a CSV file
df.to_csv('marchtee_tshirt_info.csv', index=False)

# Close the browser
driver.quit()
```

Reviews extraction code:

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import pandas as pd

# Path to your ChromeDriver executable
service = Service(executable_path="chromedriver.exe")

# Create a new instance of the Chrome driver
driver = webdriver.Chrome(service=service)

# Open the reviews page
base_url = 'https://www.marchtee.com/reviews'
driver.get(base_url)

# Wait for the page to load initially
time.sleep(30)

reviews = []

# Extract review details
review_elements = driver.find_elements(By.CSS_SELECTOR, 'div.reviews-block')
for r in review_elements:
    try:
        name = r.find_element(By.CSS_SELECTOR, 'p.sender-name').text
    except:
        name = r.find_element(By.CSS_SELECTOR, 'p.sender').text
    try:
        quote = r.find_element(By.CSS_SELECTOR, 'p.body').text
    except:
        try:
            quote = r.find_element(By.CSS_SELECTOR, 'p.comment').text
        except:
            quote = None

    reviews_info = {
        'Name': name,
        'Review': quote
    }
    reviews.append(reviews_info)

# Convert the list of dictionaries to a DataFrame
df = pd.DataFrame(reviews)

# Save the DataFrame to a CSV file
df.to_csv('marchtee_reviews.csv', index=False)


# Close the browser
driver.quit()
```
